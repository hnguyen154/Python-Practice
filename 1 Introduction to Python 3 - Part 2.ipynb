## Nested Loop

```python
b = ['banana', 'apple', 'microsoft']

for i in range(len(b)): # i = 0, 1, 2
    for j in range(i + 1):
        # i = 0 -> j = 0
        # i = 1 -> j = 0, 1
        # i = 2 -> j = 0, 1, 2
        
        print(b[i])
```

```python
n = 7

# Triangle

for i in range(n):
    for j in range(n-i):
        print(' ', end=' ')
    for k in range(2 * i + 1):
        print('*',end=' ')
    print()
    
# Pole
for i in range(n):
    for j in range(n - 1):
        print(' ', end=' ')
    print('* * *')
```

# WHILE LOOP

while (*condition*): <br>
&emsp; <i>statements</i>

```python
i = 0 

while (i < 5):
    print(i)

# Infinity Loop!!
```

```python
i = 0 

while (i < 5):
    print('Hello')
    i+=1
```

```python
i = 0
j = 0

while (i + j < 10):
    print(i,j)
    i += 1
    j += 2
```

```python
given_list = [5, 4, 4, 3, 1]

# Goal: get sum of the list

total = 0
i = 0

while i < len(given_list): 
    total += given_list[i]
    i += 1

print(total)
```

## Nested Dictionaries

```python
Student = {'Gary':  {'Mathematics': 89, 'Science': 76, 'Chemistry': 89, 'History': 76},
           'Maria': {'Mathematics': 98, 'Science': 99, 'Chemistry': 88, 'History': 84},
           'John':  {'Mathematics': 91, 'Science': 87, 'Chemistry': 77, 'History': 95}
         }
```

```python
Student
```

```python
Student['John']
```

```python
Student['Mathematics']
```

```python
Student['John']['Mathematics']
```

```python
for i in Student['John']:
    print(i)
```

```python
for i in Student:
    print(i, 'got', Student[i]['History'], '% in History')
```

```python
# Find the Max score in each subject

# First, get all the subjects 
Subject = []
for i in Student['John']:
    Subject.append(i)
```

```python
# Then, go through each subject and find the max score

for i in Subject:
    score = []
    for j in Student:
        score.append(Student[j][i])
    print(i, max(score))
```

# EXCEPTION HANDLING

**try:** <br>
&emsp;    statements/arguments <br>
**except:** <br>
&emsp;    Blank or print statement or anything else

```python
try:
    print(z)
except:
    print('Please define z')
```

```python
z = 5
```

```python
r = 'Hello'

try:
    y = float(r)
    print(y)
except:
    print('Please assign number to r')
```

```python
r = '12.53'

try:
    y = float(r)
    print(y)
except:
    print('Please assign number to r')
```

```python
r = '5646'

try:
    y = float(r)
    print(y)
except:
    print('Please assign number to r')
finally:
    print('Finished...')
```

```python
try:
    x = input('Enter Age: ')
    y = int(x) + 25
    print(y)
except:
    print('Enter Number')
finally:
    print('End of program.')
```

```python
def age():
    try:
        x = input('Enter Age: ')
        y = int(x) + 25
        print(y)
    except:
        print('Enter Number')
    finally:
        print('End of program.')
```

```python
age()
```

```python
x = 'Hello'

try:
    print(int(x))
except ValueError:
    print('Enter a number')
```

```python
x = 3
y = 0

try:
    print(x/y)
except ValueError:
    print('Division by 0 is not allowed')
```

```python
try:
    print(x/y)
except ZeroDivisionError:
    print('Division by 0 is not allowed')
```
